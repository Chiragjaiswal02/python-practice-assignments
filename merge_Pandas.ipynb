```python
import pandas as pd
import numpy as np
```

```python
sales_df = pd.DataFrame({
    'customer_id': [1, 2, 3, 1],
    'product_id': [101, 102, 103, 101],
    'quantity_sold': [2, 1, 5, 3]
})

customers_df = pd.DataFrame({
    'customer_id': [1, 2, 3],
    'customer_name': ['Alice', 'Bob', 'Charlie'],
    'customer_location': ['New York', 'Los Angeles', 'Chicago']
})

products_df = pd.DataFrame({
    'product_id': [101, 102, 103],
    'product_name': ['Laptop', 'Smartphone', 'Tablet'],
    'category': ['Electronics', 'Electronics', 'Electronics'],
    'price': [1000, 800, 500]
})
```

```python
merged_sales_customers_df = pd.merge(sales_df, customers_df, on='customer_id', how='left')
```

```python
merged_sales_customers_df
```

```python
final_df = pd.concat([merged_sales_customers_df, products_df], axis=1)
```

```python
final_df
```

```python
complete_sales_df = pd.merge(merged_sales_customers_df, products_df, on='product_id', how='left')
```

```python
complete_sales_df
```

```python
complete_sales_df['revenue'] = complete_sales_df['quantity_sold'] * complete_sales_df['price']
```

```python
complete_sales_df
```

```python
category_summary = complete_sales_df.groupby('category').agg({
    'quantity_sold': 'sum',
    'revenue': 'sum'
}).reset_index()
```
